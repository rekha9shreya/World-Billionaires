```python
# Impute missing values with median

df_world['population_country'].fillna(df_world['population_country'].median(), inplace=True)
df_world['latitude_country'].fillna(df_world['latitude_country'].median(), inplace=True)
df_world['total_tax_rate_country'].fillna(df_world['total_tax_rate_country'].median(), inplace=True)
df_world['tax_revenue_country_country'].fillna(df_world['tax_revenue_country_country'].median(), inplace=True)
df_world['life_expectancy_country'].fillna(df_world['life_expectancy_country'].median(), inplace=True)
df_world['gross_primary_education_enrollment_country'].fillna(df_world['gross_primary_education_enrollment_country'].median(), inplace=True)
df_world['gross_tertiary_education_enrollment'].fillna(df_world['gross_tertiary_education_enrollment'].median(), inplace=True)
df_world['cpi_change_country'].fillna(df_world['cpi_change_country'].median(), inplace=True)
df_world['cpi_country'].fillna(df_world['cpi_country'].median(), inplace=True)
```

```python
#rename the columns
df_world.rename(columns = {'finalWorth' : 'final_worth'}, inplace = True)
df_world.rename(columns = {'personName' : 'person_name'}, inplace = True)
df_world.rename(columns = {'countryOfCitizenship' : 'country_of_citizenship'}, inplace = True)
df_world.rename(columns = {'selfMade' : 'self_made'}, inplace = True)
df_world.rename(columns = {'birthDate' : 'birth_date'}, inplace = True)
df_world.rename(columns = {'lastName' : 'last_name'}, inplace = True)
df_world.rename(columns = {'firstName' : 'first_name'}, inplace = True)
df_world.rename(columns = {'birthYear' : 'birth_year'}, inplace = True)
df_world.rename(columns = {'birthMonth' : 'birth_month'}, inplace = True)
df_world.rename(columns = {'birthDay' : 'birth_day'}, inplace = True)
df_world.rename(columns = {'gross_tertiary_education_enrollment' : 'tertiary_edc_rate'}, inplace = True)
df_world.rename(columns = {'gross_primary_education_enrollment_country' : 'primary_edc_rate'}, inplace = True)
df_world.rename(columns = {'tax_revenue_country_country' : 'tax_revenue_country'}, inplace = True)
```

```python
# Checking for missing values in COL file
df_world.isnull().sum()
```

```python
# Impute missing values with median
df_world['birth_day'].fillna(df_world['birth_day'].median(), inplace=True)
df_world['birth_month'].fillna(df_world['birth_month'].median(), inplace=True)
df_world['birth_year'].fillna(df_world['birth_year'].median(), inplace=True)
```

```python
df_world['age'].fillna(df_world['age'].median(), inplace=True)
```

```python
# Impute missing values with Unknown
df_world['country'] = df_world['country'].fillna('Unknown')
```

```python
# Impute missing values with NaN
df_world['state'] = df_world['state'].fillna('NaN')
```

```python
# Impute missing values with NaN
df_world['city'] = df_world['city'].fillna('NaN')
df_world['birth_data'] = df_world['birth_date'].fillna('NaN')
df_world['first_name'] = df_world['first_name'].fillna('NaN')
df_world['gdp_country'] = df_world['gdp_country'].fillna('NaN')
```

```python
# Checking for missing values in COL file
df_world.isnull().sum()
```

```python
df_world['country'].value_counts(dropna = False)
```

```python
# Converting age from float to integer

df_world['age'] = df_world['age'].astype('Int64')
```

```python
# Converting birth_year, birth_day, birth_month from float to integer

df_world['birth_day'] = df_world['birth_day'].astype('Int64')
df_world['birth_month'] = df_world['birth_month'].astype('Int64')
df_world['birth_year'] = df_world['birth_year'].astype('Int64')
```
